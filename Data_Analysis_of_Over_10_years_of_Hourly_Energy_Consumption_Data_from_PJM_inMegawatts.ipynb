**Exploratory Data Analysis**

```python
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S')
```

In exploratory data analysis (EDA), we can gain insights into the relationship between daytime, month, and energy consumption in megawatts by visualizing and analyzing this data

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame with the 'EnergyConsumption' column
sns.displot(df['AEP_MW'], kde=True, bins=20, color='skyblue')
plt.title('Distribution of Energy Consumption')
plt.xlabel('Energy Consumption (MW)')
plt.ylabel('Frequency')
plt.show()
```

This is almost symmetric distribution so it represents central tendecy and we can say that from this distribution also that average enegy consumption is 15000 (aprox)

```python
plt.figure(figsize=(6, 3))
sns.lineplot(x='Year', y='AEP_MW', data=df, color='skyblue', marker='o', markersize=8, linestyle='-')
plt.title('Yearly Energy Consumption')
plt.xlabel('Year')
plt.ylabel('Total Energy Consumption (MW)')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
sns.lineplot(x=df['Time'], y=df['AEP_MW'], data=df, color='skyblue', marker='o', markersize=8, linestyle='-')
plt.title('Hourly Energy Consumption')
plt.xlabel('Hour of the Day')
plt.ylabel('Energy Consumption (MW)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.lineplot(x=df['Month'], y=df['AEP_MW'], data=df, color='skyblue', marker='o', markersize=8, linestyle='-')
plt.title('Monthly Energy Consumption')
plt.xlabel('Month of the Year')
plt.ylabel('Energy Consumption (MW)')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
